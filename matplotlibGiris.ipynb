```python
%matplotlib inline
```

```python
yas=np.arange(10,60,10)
kilo=[40,70,75,85,80]
kilo=np.array(kilo)
```

```python
plt.plot(yas,kilo,"g")
plt.xlabel("yas")
plt.ylabel("kilo")
plt.title("yaşa göre kilo değişimi")
#plt.show()
```

```python
numpyDizi1=np.linspace(1,10,20)
numpyDizi2=numpyDizi1**3
```

```python
plt.plot(numpyDizi1,numpyDizi2,"r--")
```

```python
plt.subplot(1,2,1)  #1 sıra ,2 grafik ,1.grafik
plt.plot(numpyDizi1,numpyDizi2,"b*-")
plt.subplot(1,2,2)  #1 sıra ,2 grafik ,2.grafik
plt.plot(numpyDizi2,numpyDizi1,"y-*")
```

```python
ornekfigure=plt.figure()
eksen1=ornekfigure.add_axes([0.1,0.1,0.8,0.8])
eksen2=ornekfigure.add_axes([0.3,0.5,0.3,0.3])

eksen1.plot(numpyDizi1,numpyDizi2,"r")
eksen1.set_xlabel("x eksenni")
eksen1.set_ylabel("y eksenni")
eksen1.set_title("başlık")

eksen2.plot(numpyDizi2,numpyDizi1,"b")
eksen2.set_xlabel("küçük x eksenni")
eksen2.set_ylabel("küçük y eksenni")
eksen2.set_title("küçük başlık")
```

```python
(benifigure,benimeksen)=plt.subplots(nrows=1,ncols=2)
#benimeksen.plot(numpyDizi1,numpyDizi2,"b")
for eksen in benimeksen:
    eksen.plot(numpyDizi1,numpyDizi2,"r")
    
#benime eksen array olduğu için içinde dönmek gerek
#benimfigure
plt.tight_layout()  #aralık açar daha güzel görüntü
```
